```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Метрики каждой модели
multinomial_nb_metrics = {
    'Model': 'MultinomialNB',
    'Accuracy': 0.74,
    'Precision': 0.75,
    'Recall': 0.74,
    'F1-score': 0.66
}

logistic_regression_metrics = {
    'Model': 'LogisticRegression',
    'Accuracy': 0.77,
    'Precision': 0.85,
    'Recall': 0.77,
    'F1-score': 0.71
}

random_forest_metrics = {
    'Model': 'RandomForest',
    'Accuracy': 0.78,
    'Precision': 0.84,
    'Recall': 0.78,
    'F1-score': 0.72
}

zero_shot_metrics = {
    'Model': 'ZeroShot',
    'Accuracy': 0.54,
    'Precision': 0.66,
    'Recall': 0.54,
    'F1-score': 0.56
}

# Список словарей метрик каждой модели
metrics_list = [
    multinomial_nb_metrics,
    logistic_regression_metrics,
    random_forest_metrics,
    zero_shot_metrics
]

# Датафрейм из метрик
metrics_df = pd.DataFrame(metrics_list)

# Добавление колонки по названию моделей
metrics_df.set_index('Model', inplace=True)

# Вывод датафрейма
print(metrics_df)

# Создание графика с данными каждой модели
plt.figure(figsize=(10, 6))
sns.barplot(data=metrics_df.reset_index().melt(id_vars='Model'), x='Model', y='value', hue='variable')
plt.title('Model Comparison')
plt.ylabel('Score')
plt.xlabel('Model')
plt.legend(title='Metric')
plt.show()
```
